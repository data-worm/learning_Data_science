```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine_df=pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
wine_df.columns=['class label','Alcohol','Malic acid']
```

```python
wine_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine_df.drop('class label', axis=1),     #1st parameter is x so dropeed one column
                                                    wine_df['class label'],#2nd column is the result column  
                                                    test_size=0.3,
                                                    random_state=0) 
X_train.shape, X_test.shape
```

# minmax normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```
