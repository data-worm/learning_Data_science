### remove columns that are not needed (finally we need columns that can give us tags)

```python
# needed columns = Genre,id, keywords, title (in english),overview,cast, crew. 
# As the Target was to create tags, mostly all the
# numerical columns are dropped and those columns have been kept which can help in creating tags
```

```python
df=df[['movie_id','title','overview','keywords','cast','crew','genres']]
```

```python
def find_name(input):
    L=[]
    for i in eval(input):
        L.append(i['name'])
    return L
```

```python
df['genres']=df['genres'].apply(find_name)
```

```python
df
```

```python
def find_top3_name(input):
    L=[]
    counter=0
    for i in eval(input):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
df['cast']=df['cast'].apply(find_top3_name)
```

```python
df['keywords'][0]
```

```python
df['keywords']=df['keywords'].apply(find_name)
```

```python
df['crew'][0]
```

```python
def find_dir(input):
    L=[]
    for i in eval(input):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
df['crew']=df['crew'].apply(find_dir)
```

```python
df
```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df
```

```python
#Having Blaank spaces in such a case is not recommended so we need to remove the blank spaces from all the columns
```

```python
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['tags']=df['overview']+df['genres']+df['keywords']+df['cast']+df['crew']
```

```python
df.head()
```

```python
df_final=df[['movie_id','title','tags']]
```

```python
df_final
```

```python
#join all the tags with spaces
```

```python
df_final['tags']=df_final['tags'].apply(lambda x:" ".join(x))
```

```python
df_final
```

```python
df_final['tags']=df_final['tags'].apply(lambda x:x.lower())
```

```python
df_final
```

```python
df_final['movie_id'][df_final['title'].loc[df_final['title']=='El Mariachi'].index[0]]
```

```python
#using CountVecctorizer make all the tags as vectors
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vector_of_tags=cv.fit_transform(df_final['tags']).toarray()
```

```python
vector_of_tags.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector_of_tags)
```

```python
similarity.shape
```

```python
distances= sorted(list(enumerate(similarity[2])),reverse=True,key =lambda x:x[1])
```

```python
distances
```

```python
for i in distances[1:6]:
    print(df_final.iloc[i[0]].title)
    
```

```python
def recommend_movie(input_movie_name):
    input_movie_id=df_final[df_final['title']==input_movie_name].index[0]
    distances= sorted(list(enumerate(similarity[input_movie_id])),reverse=True,key =lambda x:x[1])    
    for i in distances[1:6]:
        print(df_final.iloc[i[0]].title)
```

```python
recommend_movie('The Dark Knight Rises')
```
