```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

# 1. Relational plot
# scatter and line plot

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

##  1.1 scatter plot -> axis level function

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

### relplot -> figure level

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

## 1.2 line plot (mostly for timeseries)

```python
gap= px.data.gapminder()
```

```python
gap
```

```python
tempdf=gap[gap['country']=='India']
```

```python
tempdf
```

```python
sns.lineplot(data=tempdf,x='year',y='lifeExp',)
```

```python
sns.relplot(data=tempdf,x='year',y='lifeExp',kind='line')
```

```python
tempdf2=gap[gap['country'].isin(['India','Brazil','Germany'])]
```

```python
tempdf2
```

```python
sns.relplot(data=tempdf2,x='year',y='lifeExp',kind='line',hue='country',style='continent')
```

### Facet plot

```python
tips
```

```python
#facet plot is figure level function so it  can only work with rel plot
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='sex',col='day')
```

```python
gap
```

### column wrap

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```

# 2. Distribution plots (histplot,kdeplot,rugplot)

```python
# used for univariate
# range of observation
# central tendencies

#figure level function: displot
#axis level function:histplot,kdeplot,rugplot
```

## hist plot

```python
tips
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
#bins parameter
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

```python
#count plot
```

```python
sns.displot(data=tips,x='day',kind='hist')
```

```python
#hue parameter
```

```python
sns.displot(data=tips,x='tip',kind='hist',hue='sex')
```

```python
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.displot(data=titanic,x='age',element='step',hue='sex',kind='hist')
```

```python
sns.displot(data=titanic,x='age',element='step',col='sex',row='survived',kind='hist')
```

## KDE plot

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

## Rugplot

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

## bivariate histogram

```python
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='hist')
```

## Bivariate kde

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='kde',fill=True,height=7,aspect=2)    #important to increse size in figure level function/
```

# 3. Matrix plot

## 3.1 Heatmap  (axis level function )

```python
#rectangular data as color encoded matrix
```

```python
#what is wide and long format ?
```

```python
tempdf3=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(tempdf3)
```

```python
tempdf4=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
tempdf4
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(tempdf4,annot=True,linewidth=0.5,cmap='viridis')
```
