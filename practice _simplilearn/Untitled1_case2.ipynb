```python
df['avg_satifaction']=(df['RelationshipSatisfaction']+df['JobSatisfaction']+df['WorkLifeBalance']+df['EnvironmentSatisfaction']+df['JobInvolvement'])/5
```

```python
df['avg_satifaction']
```

```python
def satisfaction(df):
    if df['avg_satifaction']>2.5:
        return 0
    else:
        return 1
```

```python
df['satif']=df.apply(lambda df:satisfaction(df),axis=1)
```
